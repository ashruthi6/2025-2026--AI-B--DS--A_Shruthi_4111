```python
# Plotting

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("X-Y Graph")
plt.show()
```

```python
#BAR CHART
import matplotlib.pyplot as plt

# Data
subjects = ["Maths", "Physics", "Chemistry", "CS"]
marks = [80, 75, 85, 90]

# Bar chart
plt.bar(subjects, marks)

# Labels and title
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.title("Bar Chart")

# Show chart
plt.show()
```

```python
#scatter graph
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

plt.scatter(x, y)
plt.title("Scatter Graph")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
#histogram
import matplotlib.pyplot as plt

data = [10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

plt.hist(data, bins=5, edgecolor='black')
plt.title("Histogram")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt

labels = ["Maths", "Physics", "Chemistry", "CS"]
marks = [80, 75, 85, 90]

plt.pie(marks, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```
