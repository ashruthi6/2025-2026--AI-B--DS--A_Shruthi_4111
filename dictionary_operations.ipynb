```python
# 1. Creating dictionary
student = {
    "name": "Shruthi",
    "age": 20,
    "course": "B.Tech"
}

print("Original dictionary:", student)

# 2. Accessing values
print("Name:", student["name"])
print("Age using get():", student.get("age"))

# 3. Adding new key-value pair
student["marks"] = 95
print("After adding marks:", student)

# 4. Updating value
student["age"] = 21
print("After updating age:", student)

# 5. Removing elements
student.pop("course")
print("After pop:", student)

# 6. Removing last inserted item
student.popitem()
print("After popitem:", student)

# 7. Adding again for next operations
student["course"] = "B.Tech"
student["city"] = "Hyderabad"
print("After adding again:", student)

# 8. Keys, Values, Items
print("Keys:", student.keys())
print("Values:", student.values())
print("Items:", student.items())

# 9. Length of dictionary
print("Length:", len(student))

# 10. Checking key existence
print("Is 'name' present?", "name" in student)

# 11. Looping through dictionary
print("Looping through dictionary:")
for key, value in student.items():
    print(key, ":", value)

# 12. Copying dictionary
copy_dict = student.copy()
print("Copied dictionary:", copy_dict)

# 13. Clearing dictionary
temp = student.copy()
temp.clear()
print("After clear:", temp)

# 14. Nested dictionary
nested = {
    "student1": {"name": "A", "age": 20},
    "student2": {"name": "B", "age": 21}
}

print("Nested dictionary:", nested)
print("Access nested value:", nested["student1"]["name"])
```
