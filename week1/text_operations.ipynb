**Q4:Write the code to create, save and read the file from the local drive to perform the text operations ( count number of lines, words and characters)**
**DATE:03/02/2026**

```python
#  Create and write to a file
file_name = "sample_text.txt"
with open(file_name, "w") as file:
    file.write("Hello World\n")
    file.write("This is a Python file\n")
    file.write("We are counting lines, words and characters\n")
print("File created and data written successfully!")
# Read the file
with open(file_name, "r") as file:
    text = file.read()
print("\nFile Content:")
print(text)
#Count number of lines
lines = text.split("\n")
line_count = len(lines)
# Count number of words
words = text.split()
word_count = len(words)
#Count number of characters
char_count = len(text)
# Display results
print("\nNumber of lines:", line_count)
print("Number of words:", word_count)
print("Number of characters:", char_count)
```
