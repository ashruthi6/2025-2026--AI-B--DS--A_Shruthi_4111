```python
# LIST OPERATIONS

# Creating a list
lst = [1,2,3,4,5]
print(" list:", lst)

# Accessing elements
print("First element:", lst[0])
print("Last element:", lst[-1])

# Adding elements
lst.append(50)
print("After append:", lst)

lst.insert(2, 25)
print("After insert:", lst)

lst.extend([60, 70])
print("After extend:", lst)

# Removing elements
lst.remove(25)
print("After remove:", lst)

lst.pop()
print("After pop:", lst)

lst.pop(1)
print("After pop at index 1:", lst)

# Updating element
lst[0] = 100
print("After updating:", lst)

# Searching
print("Index of 3:", lst.index(3))
print("Count of 2:", lst.count(2))

# Sorting
lst.sort()
print("After sort:", lst)

lst.sort(reverse=True)
print("After reverse sort:", lst)

# Reversing
lst.reverse()
print("After reverse:", lst)

# Copying
new_lst = lst.copy()
print("Copied list:", new_lst)

# Length
print("Length of list:", len(lst))

# Membership
print("Is 40 in list?", 40 in lst)

# Looping
print("Elements in list:")
for i in lst:
    print(i)

# List comprehension
squares = [i*i for i in lst]
print("Squares of elements:", squares)

# Clearing list
lst.clear()
print("After clear:", lst)
```
