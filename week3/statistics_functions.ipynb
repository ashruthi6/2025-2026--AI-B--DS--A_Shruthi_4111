**Q14:Read any of the data set from URL and perform the descriptive statistics functions**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Read dataset from URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

print("Dataset:")
df = pd.read_csv(url)
print(df.head())

# Select one column for statistics
data = df['sepal_length']

# Summary statistics
print("\nSummary statistics:")
print(data.describe())

# Variance and Standard deviation
print("\nVariance:", np.var(data))
print("Standard deviation:", np.std(data))

# Mean, Median, Mode
mean_value = np.mean(data)
median_value = np.median(data)
mode_value = data.mode()

# Histogram
plt.hist(data, bins=10, color='skyblue')

# Mean line
plt.axvline(mean_value, color='red', linewidth=2, label=f"Mean: {mean_value:.2f}")

# Median line
plt.axvline(median_value, color='green', linewidth=2, label=f"Median: {median_value:.2f}")

plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.title('Distribution of Sepal Length')
plt.legend()
plt.show()

# Print Mean, Median, Mode
print("\nInference:")
print("Mean:", mean_value)
print("Median:", median_value)

if len(mode_value) == 1:
    print("Mode:", mode_value[0])
else:
    print("Mode:", list(mode_value))

# Distribution inference
if mean_value > median_value:
    print("The distribution is positively skewed")
elif mean_value < median_value:
    print("The distribution is negatively skewed")
else:
    print("The distribution is symmetric")
```
