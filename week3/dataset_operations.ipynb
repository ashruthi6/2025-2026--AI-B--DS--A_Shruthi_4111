**Q12**:Write the code to implement the following operations on any data set reading from URL
**Data filtering**
**Data Manipulation**
**Data ranking**

```python
# Import required library
import pandas as pd

# Step 1: Read dataset from URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

# Display original dataset
print("Original Dataset:")
print(df.head())

# -----------------------------------
# 1. Data Filtering
# -----------------------------------

# Filter records where total_bill > 20
filtered_data = df[df['total_bill'] > 20]
print("\nFiltered Data (total_bill > 20):")
print(filtered_data.head())

# Filter records where day is Sunday
sunday_data = df[df['day'] == 'Sun']
print("\nFiltered Data (day = Sunday):")
print(sunday_data.head())

# -----------------------------------
# 2. Data Manipulation
# -----------------------------------

# Create new column tip_percentage
df['tip_percentage'] = (df['tip'] / df['total_bill']) * 100

print("\nDataset after adding tip_percentage column:")
print(df.head())

# Calculate average total_bill
avg_bill = df['total_bill'].mean()
print("\nAverage Total Bill:", avg_bill)

# -----------------------------------
# 3. Data Ranking
# -----------------------------------

# Rank based on total_bill
df['Rank_total_bill'] = df['total_bill'].rank()

print("\nDataset with Ranking based on total_bill:")
print(df[['total_bill', 'Rank_total_bill']].head())

# Rank based on tip in descending order
df['Rank_tip'] = df['tip'].rank(ascending=False)

print("\nDataset with Ranking based on tip (Descending):")
print(df[['tip', 'Rank_tip']].head())
```
