```python
# 1. Creating sets
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7, 8}
empty_set = set()   # empty set (not {})

print("Set s1:", s1)
print("Set s2:", s2)
print("Empty set:", empty_set)

# 2. Adding elements
s1.add(10)
print("After add:", s1)

# 3. Adding multiple elements
s1.update([11, 12])
print("After update:", s1)

# 4. Removing elements
s1.remove(12)   # error if element not present
print("After remove:", s1)

# 5. Discard (no error if element not present)
s1.discard(20)
print("After discard:", s1)

# 6. Pop element (removes random element)
popped = s1.pop()
print("After pop:", s1, "| Popped:", popped)

# 7. Union (combine sets)
print("Union:", s1.union(s2))

# 8. Intersection (common elements)
print("Intersection:", s1.intersection(s2))

# 9. Difference (elements in s1 not in s2)
print("Difference s1 - s2:", s1.difference(s2))

# 10. Symmetric difference (not common elements)
print("Symmetric difference:", s1.symmetric_difference(s2))

# 11. Membership test
print("Is 3 present in s1?", 3 in s1)

# 12. Length of set
print("Length of s1:", len(s1))

# 13. Looping through set
print("Looping through s1:")
for item in s1:
    print(item)

# 14. Copying set
copy_set = s1.copy()
print("Copied set:", copy_set)

# 15. Clearing set
temp = s1.copy()
temp.clear()
print("After clear:", temp)
```
