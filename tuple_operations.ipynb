```python
# 1. Creating tuples
t1 = (1, 2, 3, 4, 5)
t2 = ("apple", "banana", "cherry")
single = (10,)   # single element tuple (comma compulsory)

print("Tuple t1:", t1)
print("Tuple t2:", t2)
print("Single element tuple:", single)

# 2. Accessing elements
print("First element:", t1[0])
print("Last element:", t1[-1])

# 3. Slicing
print("Slice [1:4]:", t1[1:4])
print("Slice [:3]:", t1[:3])
print("Slice [::2]:", t1[::2])

# 4. Tuple length
print("Length of tuple:", len(t1))

# 5. Concatenation (joining tuples)
t3 = t1 + t2
print("Concatenated tuple:", t3)

# 6. Repetition
t4 = t1 * 2
print("Repeated tuple:", t4)

# 7. Membership test
print("Is 3 present?", 3 in t1)
print("Is 10 present?", 10 in t1)

# 8. Iteration (looping)
print("Looping through tuple:")
for item in t1:
    print(item)

# 9. Built-in functions
print("Minimum:", min(t1))
print("Maximum:", max(t1))
print("Sum:", sum(t1))

# 10. Counting elements
t5 = (1, 2, 3, 2, 2, 4)
print("Count of 2:", t5.count(2))

# 11. Finding index
print("Index of 3:", t1.index(3))

# 12. Converting tuple to list (to modify)
temp = list(t1)
temp.append(6)
t1 = tuple(temp)
print("After adding element:", t1)

# 13. Nested tuple
nested = ((1, 2), (3, 4), (5, 6))
print("Nested tuple:", nested)
print("Access nested element:", nested[0][1])
```
