#18)Implement the data cleaning operations on any data set using URL

```python

import pandas as pd
import numpy as np

u = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"

d = pd.read_csv(u)

print("=== Original Dataset Info ===")
print(d.info())

print("\nMissing Values Before Cleaning:")
print(d.isnull().sum())

print("\nFirst 5 rows:")
print(d.head())

d = d.drop_duplicates()

d["age"] = d["age"].fillna(d["age"].mean())
d["embarked"] = d["embarked"].fillna(d["embarked"].mode()[0])

print("\n=== Missing Values After Filling ===")
print(d.isnull().sum())

if "deck" in d.columns:
    d = d.drop(columns=["deck"])

print("\n=== Columns After Dropping 'deck' ===")
print(d.columns)

d["survived"] = d["survived"].astype("category")
d["pclass"] = d["pclass"].astype("category")

print("\n=== Data Types After Conversion ===")
print(d.dtypes)

d.columns = d.columns.str.lower()

print("\n=== Columns After Renaming ===")
print(d.columns)

q1 = d["age"].quantile(0.25)
q3 = d["age"].quantile(0.75)
iqr = q3 - q1

low = q1 - 1.5 * iqr
high = q3 + 1.5 * iqr

b = len(d)
d = d[(d["age"] >= low) & (d["age"] <= high)]
a = len(d)

print("\nRemoved", b - a, "outlier rows based on Age")

d.to_csv("cleaned_titanic.csv", index=False)

print("\nData Cleaning Completed Successfully!")
print("Cleaned file saved as: cleaned_titanic.csv")

print("\n=== First 5 rows of Cleaned Dataset ===")
print(d.head())
```
