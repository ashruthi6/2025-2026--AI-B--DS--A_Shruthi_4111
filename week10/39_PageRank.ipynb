```python
import networkx as nx
G = nx.DiGraph()
G.add_edges_from([
    ('A', 'B'),
    ('A', 'C'),
    ('B', 'C'),
    ('C', 'A'),
    ('D', 'C')
])
pr = nx.pagerank(G)
print("PageRank Scores:")
for node, score in pr.items():
    print(node, ":", score)
```
