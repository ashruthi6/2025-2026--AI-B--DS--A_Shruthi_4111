```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.Graph()
G.add_edges_from([('A','B'),('A','C'),('B','C'),('B','D'),('C','E'),('D','E')])
ego=nx.ego_graph(G,'B')
nx.draw(ego,with_labels=True,node_color="lightblue",node_size=1000)
plt.show()
print("nodes in ego network:",ego.nodes())
print("edges in ego network:",ego.edges())
```
