#24)write a code to implement linear regression with an example URL



```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

url = "https://raw.githubusercontent.com/ybifoundation/Dataset/main/Salary%20Data.csv"
data = pd.read_csv(url)
print("Dataset Preview")
print(data.head())
X = data[['Experience Years']]
response = data['Salary']
model = LinearRegression()
model.fit(X, response)
print("\nModel Parameters")
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
years = [[5]]
predicted_salary = model.predict(years)
print("\nPredicted Salary for 5 Years Experience:", predicted_salary[0])
y_pred = model.predict(X)
mse = mean_squared_error(response, y_pred)
print("\nMean Squared Error:", mse)
plt.scatter(X, response)
plt.plot(X, y_pred)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression Salary Prediction")
plt.show()
```
