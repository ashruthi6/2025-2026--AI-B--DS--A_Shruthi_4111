**Q13**:write the code to implementthe following descriptive statistics operations any of the created data frame
**A.mean,median and mode**
**B.standard deviations**
**C.variance**
**D.kurtosis**
**E.skewness**

```python
import pandas as pd
import numpy as np
# Create a sample DataFrame
data = {
    'Marks': [45, 56, 67, 45],
    'Age': [18, 19, 20, 18]
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
# 1. Mean, Median, Mode
print("\nMean:")
print(df.mean())
print("\nMedian:")
print(df.median())
print("\nMode:")
print(df.mode())
# 2. Standard Deviation
print("\nStandard Deviation:")
print(df.std())
# 3. Variance
print("\nVariance:")
print(df.var())
# 4. Kurtosis
print("\nKurtosis:")
print(df.kurt())
# 5. Skewness
print("\nSkewness:")
print(df.skew())
```
